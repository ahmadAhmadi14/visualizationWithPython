```python
import pandas as pd
import yfinance as yf
```

```python
tickers = ["BBCA.JK","TLKM.JK","INCO.JK"]
```

```python
stock_data = yf.download(tickers, start = "2017-01-01", end = "2022-07-31")
```

```python
import matplotlib.pyplot as matplt
data = stock_data.loc[:,"Open"].copy()
data.plot(figsize=(25,7), fontsize = 17)
matplt.style.use("seaborn")
matplt.show()
```
